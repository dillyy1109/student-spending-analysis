## Feature Engineering

```python
spending_cols = [
    "housing", "food", "transportation", "books_supplies",
    "entertainment", "personal_care", "technology",
    "health_wellness", "miscellaneous"
]
df["total_spending"] = df[spending_cols].sum(axis=1)
df[["monthly_income", "total_spending"]].head()
```

```python
df["spending_status"] = df.apply(
    lambda row: "Overspending" if row["total_spending"] > row["monthly_income"]
    else "Underspending",
    axis=1
)
df["spending_status"].value_counts()
```

## Insights
- To better analyze student spending behavior, new variables were created from the existing data. First, a new feature called `total_spending` was calculated by summing all individual spending categories, including housing, food, transportation, and other expenses. This provides a comprehensive measure of each student’s overall spending.
- In addition, a new categorical variable, `spending_status`, was created to classify students based on their financial behavior. Students were labeled as “Overspending” if their total spending exceeded their monthly income, and “Underspending” otherwise.
- The results show that the vast majority of students fall into the overspending category, indicating that most students spend more than their monthly income. This suggests that students may rely on additional financial support, such as financial aid or savings, to cover their expenses.
- These engineered features allow for a deeper analysis of spending patterns and provide a clearer understanding of student financial behavior.

- First, the average total spending was compared across gender groups. The results show that spending levels are very similar among male, female, and non-binary students, with only minimal differences. This suggests that gender does not play a significant role in influencing overall spending behavior in this dataset.
- Next, the relationship between monthly income and total spending was analyzed using correlation. The correlation coefficient is approximately 0.0015, indicating almost no linear relationship between income and spending. This suggests that higher income does not necessarily lead to higher spending among students, and that other factors may influence financial behavior.
- Spending patterns were also compared across different preferred payment methods. Students using credit or debit cards have slightly higher average spending compared to those using cash or mobile payment applications. However, the differences are relatively small, suggesting that payment method has only a limited impact on total spending.
- Finally, students were grouped into spending levels (Low, Medium, High) to examine how income, financial aid, and age vary across these categories. The results show minimal differences in age across spending groups, indicating that age is not a strong factor influencing spending behavior. Similar comparisons for income and financial aid provide additional context for understanding how students manage their finances.
- Overall, the analysis suggests that student spending behavior is relatively consistent across demographic groups, with only minor variations influenced by payment methods and financial characteristics.

## Visualization

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
avg_spending.plot(kind="bar")
plt.title("Average Spending by Category")
plt.xlabel("Category")
plt.ylabel("Avererage Amount Spent")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
status_percent = df["spending_status"].value_counts(normalize=True) * 100
status_percent
status_percent.plot(kind="bar")
plt.title("Student Spending Behavior (Over vs Under)")
plt.ylabel("Percentage (%)")
plt.xlabel("Spending Status")
plt.xticks(rotation=0)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8,5))
sns.histplot(df["total_spending"], bins=30, kde=True)
plt.title("Distribution of Total Student Spending")
plt.xlabel("Total Spending")
plt.ylabel("Number of Students")
plt.tight_layout()
plt.show()
```

```python
year_spending = df.groupby("year_in_school")["total_spending"].mean().sort_values()
year_spending
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="year_in_school", y="total_spending", estimator="mean")
plt.title("Average Total Spending by Year in School")
plt.xlabel("Year in School")
plt.ylabel("Average Total Spending")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="gender", y="total_spending", estimator="mean")
plt.title("Average Total Spending by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Total Spending")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.barplot(data=df, x="preferred_payment_method", y="total_spending", estimator="mean")
plt.title("Average Total Spending by Preferred Payment Method")
plt.xlabel("Preferred Payment Method")
plt.ylabel("Average Total Spending")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```
